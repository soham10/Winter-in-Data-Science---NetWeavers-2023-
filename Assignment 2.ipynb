```python
import numpy as np 
import matplotlib.pyplot as plt

# Fonts for titles and axis labels
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'darkred','size':20}
font3 = {'family':'fantasy','color':'black','size':25}

# Training Data
x_train=np.array([3,5,8])
y_train=np.array([4,7,11])

# A
y=(7/5)*x_train
plt.scatter(x_train,y_train)
plt.plot(x_train,y,c='r')
plt.title('Employee Database',fontdict=font3)
plt.xlabel('No. of years experience',fontdict=font1)
plt.ylabel('Salary',fontdict=font2)
plt.show()

# B
def compute_cost(w,b):
    J=0
    for i in range(len(x_train)):
        J=J+(w*x_train[i]+b-y_train[i])**2
    print(f'Cost Function,J={J/(2*len(x_train))}')

# C
def compute_gradient(w,b):
    A=0
    B=0
    for i in range(len(x_train)):
        A=A+(w*x_train[i]+b-y_train[i])*x_train[i]
        B=B+(w*x_train[i]+b-y_train[i])
    return A/len(x_train),B/len(x_train)
```

Q2

```python
def compute_gradient_descent(x_train,y_train,w_init,b_init,alpha,num_iters):
    w=w_init
    b=b_init
    for i in range(num_iters):
        new_w=w-alpha*compute_gradient(w,b)[0]
        new_b=b-alpha*compute_gradient(w,b)[1]
        w=new_w
        b=new_b
    return w,b

tup_w_b=compute_gradient_descent(x_train,y_train,0,0,0.01,10000)
print(f'final w={tup_w_b[0]}\nfinal b={tup_w_b[1]}')
compute_cost(tup_w_b[0],tup_w_b[1])

y=tup_w_b[0]*x_train+tup_w_b[1]
plt.scatter(x_train,y_train)
plt.plot(x_train,y,marker='o',c='r')
plt.title('Employee Database',fontdict=font3)
plt.xlabel('No. of years experience',fontdict=font1)
plt.ylabel('Salary',fontdict=font2)
plt.show()
```
